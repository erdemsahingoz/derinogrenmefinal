```
import os

# İndirilen zip dosyasının klasör yapısına göre ana dizin yolunu belirliyoruz
base_path = '/content/datasets'
if os.path.exists('/content/datasets/brain-tumor'):
    base_path = '/content/datasets/brain-tumor'

# YAML dosyasının içeriği
yaml_content = f"""
path: {base_path}
train: train/images
val: valid/images
nc: 2
names:
  0: negatif
  1: pozitif
"""

# Dosyayı oluşturma ve yazma
with open('/content/brain-tumor.yaml', 'w', encoding='utf-8') as f:
    f.write(yaml_content.strip())

print("brain-tumor.yaml dosyası başarıyla oluşturuldu!")
```

```
# Model eğitimini başlatıyoruz
!yolo detect train data=/content/brain-tumor.yaml model=yolov8n.pt epochs=50 imgsz=640
```

```
# YAML içeriğini dosya yapına göre güncelliyoruz
yaml_content = """
path: /content/datasets
train: images/train
val: images/val
nc: 2
names:
  0: negatif
  1: pozitif
"""

with open('/content/brain-tumor.yaml', 'w', encoding='utf-8') as f:
    f.write(yaml_content.strip())

print("YAML dosyası doğru yollarla güncellendi!")
```
